## Penggabungan Data

```python
import pandas as pd
import glob
import os

# Folder tempat file CSV disimpan
folder_path = "data"  # ubah sesuai lokasi folder kamu

# Ambil semua file CSV
all_files = glob.glob(os.path.join(folder_path, "*.csv"))

# List untuk menampung dataframe
dataframes = []

for file in all_files:
    # Ambil nama wisata dari nama file
    nama_wisata = os.path.splitext(os.path.basename(file))[0]

    # Baca file CSV
    df = pd.read_csv(file)

    # Tambahkan kolom Nama_Wisata di depan
    df.insert(0, "Nama_Wisata", nama_wisata)

    # Masukkan ke list
    dataframes.append(df)

# Gabungkan semua dataframe
gabungan = pd.concat(dataframes, ignore_index=True)

# Simpan ke file baru
gabungan.to_csv("gabungan_wisata.csv", index=False)

print("âœ… Semua data berhasil digabung dan kolom 'Nama_Wisata' ada di depan.")
```

## Missing Value Dan Duplicate data

```python
import pandas as pd

# Ganti dengan nama file CSV kamu
filename = "gabungan_wisata.csv"

# Membaca file CSV
df = pd.read_csv(filename)

print("=== Analisis Data CSV ===\n")

# ğŸ” Jumlah duplikat (berdasarkan semua kolom)
total_duplikat = df.duplicated().sum()
print(f"Jumlah data duplikat (semua kolom): {total_duplikat}")

# Jika kamu ingin cek duplikat berdasarkan kolom 'Review'
if 'Review' in df.columns:
    duplikat_review = df.duplicated(subset=['Review']).sum()
    print(f"Jumlah duplikat berdasarkan kolom 'Review': {duplikat_review}")

# âŒ Jumlah missing value (NaN) per kolom
missing_per_column = df.isnull().sum()

print("\n=== Missing Value per Kolom ===")
print(missing_per_column)

# ğŸ”¢ Total missing value keseluruhan
total_missing = missing_per_column.sum()
print(f"\nTotal missing value di seluruh dataset: {total_missing}")
```
